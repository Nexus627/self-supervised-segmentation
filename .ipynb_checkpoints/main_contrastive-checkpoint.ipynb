```python
import torch
from pixel_level_contrastive_learning import PixelCL
from torchvision import models
import os
import numpy as np
import time
import datetime
from PIL import Image
from skimage import io, img_as_ubyte, exposure
from glob import glob
from matplotlib import pyplot as plt

import torchvision.transforms as transforms

from torch.utils.data import DataLoader
from torch.utils.data import Dataset

resnet = models.resnet50(pretrained=True)

learner = PixelCL(
    resnet,
    image_size = 256,
    hidden_layer_pixel = 'layer4',  # leads to output of 8x8 feature map for pixel-level learning
    hidden_layer_instance = -2,     # leads to output for instance-level learning
    projection_size = 256,          # size of projection output, 256 was used in the paper
    projection_hidden_size = 2048,  # size of projection hidden dimension, paper used 2048
    moving_average_decay = 0.99,    # exponential moving average decay of target encoder
    ppm_num_layers = 1,             # number of layers for transform function in the pixel propagation module, 1 was optimal
    ppm_gamma = 2,                  # sharpness of the similarity in the pixel propagation module, already at optimal value of 2
    distance_thres = 0.7,           # ideal value is 0.7, as indicated in the paper, which makes the assumption of each feature map's pixel diagonal distance to be 1 (still unclear)
    similarity_temperature = 0.3,   # temperature for the cosine similarity for the pixel contrastive loss
    alpha = 1.,                      # weight of the pixel propagation loss (pixpro) vs pixel CL loss
    use_pixpro = True,               # do pixel pro instead of pixel contrast loss, defaults to pixpro, since it is the best one
    cutout_ratio_range = (0.6, 0.8)  # a random ratio is selected from this range for the random cutout
).cuda()

opt = torch.optim.Adam(learner.parameters(), lr=1e-4)
```

```python
class ImageDataset(Dataset):
    def __init__(self, root, transform=None, re_norm=False):
        self.re_norm = re_norm
        self.transform = transforms.Compose(transform)
        self.files = sorted(glob(root+"/*.*"))

    def __getitem__(self, index):
        img = io.imread(self.files[index])
        ### duplicate channel in order for colorjittering to work
        img = np.repeat(img[:, :, np.newaxis, ], 3, axis=2)
        # print(img.shape)
        if self.re_norm:
            img = exposure.rescale_intensity(img, in_range=(np.percentile(img, 0), np.percentile(img, 100)), out_range=(0, 1))
        img = Image.fromarray(img_as_ubyte(img))
        img= self.transform(img)
        return img
    def __len__(self):
        return len(self.files)
```

```python
def plot_loss_vs_epochs(epoch_train_loss, epoch_test_loss, fig_title):
    '''
    Plotting train and test error vs no of epochs
    '''
    epoch = np.arange(0, len(epoch_train_loss))
    plt.plot(epoch, epoch_train_loss, label='Training Loss')
    plt.plot(epoch, epoch_test_loss, label='Validation Loss')
    plt.title(fig_title)
    plt.xlabel("Epochs")
    plt.ylabel("Loss")
    plt.legend(loc='upper right')
    plt.show()
```

```python
dataset_folder="dataset/train"
val_dataset_folder="dataset/val"
batch_size=8
n_epochs=100
img_size=256
log_int=5
min_loss = 100000

transform = [
    transforms.RandomAffine(180, scale=(0.75, 1.5), shear=45),
    transforms.RandomCrop(img_size),
    transforms.ToTensor()]
dataset = ImageDataset(root=dataset_folder, transform=transform)
train_dataloader = DataLoader(dataset, batch_size=batch_size,
                                         shuffle=True, num_workers=0)

val_dataset = ImageDataset(root=val_dataset_folder, transform=transform)
val_dataloader = DataLoader(val_dataset, batch_size=batch_size,
                                         shuffle=True, num_workers=0)

prev_time = time.time()

train_loss_list, val_loss_list = [], []

for epoch in range(0, n_epochs):
    train_loss = 0
    iter_count = 0
    val_loss = 0
    for i, batch in enumerate(train_dataloader):
        images = batch.cuda()
        loss = learner(images) # if positive pixel pairs is equal to zero, the loss is equal to the instance level loss

        opt.zero_grad()
        loss.backward()
        opt.step()
        learner.update_moving_average() # update moving average of target encoder
        batches_done = epoch * len(train_dataloader) + i
        batches_left = n_epochs * len(train_dataloader) - batches_done
        time_left = datetime.timedelta(seconds=batches_left * (time.time() - prev_time))
        prev_time = time.time()
        train_loss += loss.item()
        iter_count += 1
        if i%log_int==0:
            loss, positive_pairs = learner(images, return_positive_pairs = True)
            print("[Epoch %d/%d] [Batch %d/%d] [loss: %f] [pos pairs: %d] ETA: %s"
            % (
                epoch,
                n_epochs,
                i,
                len(train_dataloader),
                train_loss/iter_count,
                positive_pairs,
                time_left,
                )
            )
            
    for i, batch in enumerate(val_dataloader):
        images = batch.cuda()
        loss = learner(images)
        val_loss += loss.item()
    
    train_loss_list.append(train_loss/len(train_dataloader))
    val_loss_list.append(val_loss/len(val_dataloader))
        
    if (val_loss/len(val_dataloader)) < min_loss:
        min_loss = val_loss/len(val_dataloader)
        print('---------------------Model saved------------------------')
        torch.save(resnet, 'improved-resnet.pt')
```

```python
plot_loss_vs_epochs(train_loss_list, val_loss_list, 'Loss')
```
